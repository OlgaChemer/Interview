2. Объединить (конкатенация) несколько файлов Excel в одну таблицу. Файлы в папке 1.

```python
path = r'/Users/olgakamskaa/Sber_school'
path = os.getcwd()
name_files = '1_sp500'

files = []
df = pd.DataFrame({'file_list':os.listdir(path)})
df
for i in df['file_list']:
    if (i.find('.xlsx') >= 0) & (i.find(name_files) >= 0):
        files.append(i)

df = pd.concat([pd.read_excel(f) for f in files], ignore_index=True)

with open('final.xlsx', 'w') as final:
    df.to_excel('/Users/olgakamskaa/Sber_school/final.xlsx')
```

```python
df_concat = pd.read_excel('final.xlsx')
df_concat.head(5)
```

3. "Смержить" (аналог JOIN в SQL ) таблицу получившуюся на предыдущем шаге с данными в файле 2_sp500. Файл в папке 2.

```python
sp2 = pd.read_excel('2_sp500.xlsx')
df_merge = df_concat.merge(sp2, on='Symbol' ,how='left')
```

```python
df_merge.head(5)
```

4. Построить сводную таблицу. Средний прайс по секторам

```python
df_avg_by_sector = df_merge.groupby('Sector')['Price'].mean().sort_values(ascending=False).reset_index()
df_avg_by_sector
```
